### Figures
High level container which contains axes

```python
# instantiate figure
fig = plt.figure()

# add axes to the figure
ax = fig.add_axes([0, 0, 1, 1])
# first arg 0: position of left side of axes wrt figure
# second arg 0: position of bottom of axes wrt figure
# third arg 1: width of axes wrt figure (1 means cover whole figure width)
# fourth arg 1: height of axes wrt figure

plt.show()
```

## Subplot2Grid function
When you want to further customize axes position

```python
# First tuple creates grid, second tuple points position of axes in grid

ax1 = plt.subplot2grid((2,3),(0,0), rowspan=2)

ax2 = plt.subplot2grid((2,3),(0,1))

ax3 = plt.subplot2grid((2,3),(0,2))
ax3.yaxis.tick_right()

ax4 = plt.subplot2grid((2,3),(1,1), colspan=2)
ax4.yaxis.tick_right()
```

## Matplotlib Patches
Matplotlib has special module called 'patches', which allows to draw shapes (e.g. rectangle, circle, complex shapes)
on graphs

```python
fix, ax = plt.subplots()

ax.add_patch(
    matplotlib.patches.Rectangle(
        (0.1,0.1),
        0.5,
        0.5,
        fill=True,
        facecolor='y',
        edgecolor='g',
        hatch='.',
        linestyle='dashed'
    )
)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for patch in [
    matplotlib.patches.Circle(
        (0.5, 0.35), 0.3,
        facecolor='red',
        hatch='*'
    ),
    
    matplotlib.patches.Circle(
        (0.5, 0.75), 0.2,
        facecolor='green',
        hatch='//'
    
    )
]:
    ax.add_patch(patch)
    
plt.show()
```

### Drawing Paths & Bezier curves

```python
import matplotlib.patches as patches
from matplotlib.path import Path
```

```python
fig, ax = plt.subplots()

p = patches.PathPatch(Path([(0.1, 0.1),    (0.8, 0.8),   (0.8, 0.1),  (0.4, 0.2)],
                              [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO]),
                     fill=None)

ax.add_patch(p)

plt.show()
```

```python
fig, ax = plt.subplots()

p = patches.PathPatch(Path([(0.1, 0.1),    (0.8, 0.8),   (0.8, 0.1),  (0.4, 0.2)],
                              [Path.MOVETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]),
                     fill=None)

ax.add_patch(p)

plt.show()
```

```python
fig, ax = plt.subplots()

p = patches.PathPatch(Path([(0.1, 0.1),    (0.8, 0.8),   (0.8, 0.1),  (0.4, 0.2)],
                              [Path.MOVETO, Path.LINETO, Path.CURVE3, Path.CURVE3]),
                     fill=None)

ax.add_patch(p)

plt.show()
```

#### converting box plot to patches to gain fine controls

```python
bp = plt.boxplot(x, vert=False, patch_artist=True)
bp
```

box plot patch is a dictionary which allows to access components of 2D patch

```python
print('number of Boxes :', len(bp['boxes']))
print('number of whiskers :', len(bp['whiskers']))
print('number of caps :', len(bp['caps']))
print('number of medians :', len(bp['medians']))
print('number of fliers :', len(bp['fliers']))
```

```python
bp = plt.boxplot(x, vert=False, patch_artist=True)

bp['boxes'][0].set(facecolor='lightyellow',
                  edgecolor='y',
                  hatch='.')
```

```python
bp = plt.boxplot(x, vert=False, patch_artist=True)

bp['boxes'][0].set(facecolor='lightyellow',
                  edgecolor='y',
                  hatch='.')

bp['whiskers'][0].set(color='red', linewidth=2)
bp['whiskers'][1].set(color='red')

bp['caps'][0].set(color='red', linewidth=2)
bp['caps'][1].set(color='red')

bp['fliers'][0].set(marker='s', markerfacecolor='g')
```

```python
array = np.random.rand(20,3)*100
array
```

```python
bp = plt.boxplot(array, patch_artist=True)
bp
# plt.show()
```

```python
print('number of Boxes :', len(bp['boxes']))
print('number of whiskers :', len(bp['whiskers']))
print('number of caps :', len(bp['caps']))
print('number of medians :', len(bp['medians']))
print('number of fliers :', len(bp['fliers']))
```

```python
colors=['blue', 'grey', 'lawngreen']

bp = plt.boxplot(array, patch_artist=True)

for i in range(len(bp['boxes'])):
    bp['boxes'][i].set(facecolor=colors[i])
    bp['whiskers'][2*i].set(color=colors[i], linewidth=3)
    bp['whiskers'][2*i + 1].set(color=colors[i], linewidth=3)
    
plt.xticks([1,2,3],['speech','reading','writing'])

plt.show()
```

```python
vp = plt.violinplot(array, showmedians=True)
plt.xticks([1,2,3],['speech','reading','writing'])

plt.show()
```

```python
vp
```

```python
vp = plt.violinplot(array, showmedians=True, vert=False)
plt.yticks([1,2,3],['speech','reading','writing'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i])

plt.show()
```

## Pi Charts

```python
sectors = ['finance', 'banking', 'crypto', 'retail', 'wealth']
pctg = [20, 30, 5, 15, 30]


fig = plt.figure(figsize=(6,6))
plt.pie(pctg, labels=sectors)

# by default, first wedge starts at 3 o'clock position & moves counter clockwise
```

```python
fig = plt.figure(figsize=(8,8))

colours = ['r', 'b', 'g', 'y', 'lime']
plt.pie(pctg, labels=sectors, colors=colours, autopct='%.2f')

plt.show()
```
