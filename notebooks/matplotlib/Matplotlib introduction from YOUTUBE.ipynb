## Pie Charts

```python
plt.style.use('ggplot')

slices=[40, 60]
labels=['Label1', 'Label2']

plt.pie(slices, labels=labels)
```

```python
wedges, texts = plt.pie(slices, labels=labels)
print('Patch :-', wedges)
print('Text :-', texts)

wedges[0].set(edgecolor='k', hatch='*')
```

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
# Explode Label4 out of pie chart

slices=[40, 60, 120, 20, 200]
labels=['Label1', 'Label2', 'Label3', 'Label4', 'Label5']

explode=[0,0,0,0.2,0]

plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'}, autopct='%0.1f')

plt.show()
```

## Stack/Area plots

```python
fig = plt.figure(figsize=(8,4))

minutes = np.arange(9)+1

player1Goals = [0,0,1,1,2,3,4,4,6]
player2Goals = [0,2,3,5,7,7,8,8,9]
player3Goals = [0,3,5,6,7,8,8,8,8]

labels=['player1','player2','player3']

plt.stackplot(minutes, player1Goals, player2Goals, player2Goals, labels=labels)

plt.legend(loc='upper left')
plt.show()
```

## Histograms

```python
plt.style.use('fivethirtyeight')

ages = [18,19,20,23,25,28,20,30,35,36,34,40,43,44,56,61]
bins = [10,20,30,40,50,60]

plt.hist(ages, bins=bins, edgecolor='k')

median_age=29
plt.axvline(median_age, color='y', label='Median Age')

plt.title('Age of Respondents')
plt.xlabel('Age')
plt.legend()
plt.show()
```

## Scatter plots

```python
plt.style.use('seaborn')

x = np.random.randint(low=1, high=10, size=50)
y = np.random.randint(low=1, high=10, size=50)

plt.scatter(x, y, marker='*')

plt.show()
```

```python
colors = np.random.randint(low=1, high=10, size=50)

plt.scatter(x, y, s=100, edgecolor='k', c=colors, cmap='Greens', alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.show()
```

```python
x = np.random.randint(low=1, high=10, size=50)
y = np.random.randint(low=1, high=10, size=50)
satisfaction = np.random.randint(low=1, high=10, size=50)
sizes = np.random.randint(low=100, high=1000, size=50)

plt.scatter(x, y, s=sizes, edgecolor='k', c=satisfaction, cmap='Greens', alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.show()
```

## Plotting Time-series Data

```python
from datetime import datetime, timedelta
from matplotlib import dates as mplt_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 31),
]

y = np.random.randint(low=3, high=12, size=7)
```

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()  #rotates dates on x-axis; gcf=Get Current Figure

plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()  #rotates dates on x-axis

date_format = mplt_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)  #gca = Get Current Axis

plt.tight_layout()
plt.show()
```
